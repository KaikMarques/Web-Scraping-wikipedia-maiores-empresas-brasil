## Pandas read_html

#### link:  https://pt.wikipedia.org/wiki/Lista_das_maiores_empresas_do_Brasil

```python
import pandas as pd
URL = 'https://pt.wikipedia.org/wiki/Lista_das_maiores_empresas_do_Brasil'

tabelas = pd.read_html(URL)
tabelas
```

```python
len(tabelas)
```

```python
tabelas[0]
```

```python
# Removendo o indice da coluna "#"

df_forbes_2000 = tabelas[0]
df_forbes_2000 = df_forbes_2000.drop(columns=["#"])
df_forbes_2000
```

## Trabalhando com dados

```python
# Renomeando as colunas usando o método str.replace

df_forbes_2000.columns = df_forbes_2000.columns.str.replace(r" (em bilhões de US$)", "")
df_forbes_2000.columns
```
